```python
from core.utils import Tibanna
from core import ff_utils


# for a given experiment set and some parameters like instrument
# print set of files and their partA hic workflow status
# if there are one that are running report the number of running cases
# if there are file pairs that don't have a corresponding part A, report them separately

env = 'fourfront-webprod'
tibanna = Tibanna(env=env)
ff = ff_utils.fdn_connection(key=tibanna.ff_keys)
pub_exps = ff_utils.get_metadata('1d9ba65e-4fb5-445b-9d7f-7c2e8e3c9a8a', connection=ff)['exp_sets_prod_in_pub']

```

```python
for exps in pub_exps:
    print "EXPERIMENT_SET", exps    
    sets = ff_utils.get_metadata(exps, connection=ff)['experiments_in_set']
    for exp in sets:
        print "EXPERIMENT", exp
        files = ff_utils.get_metadata(exp, connection=ff, frame='embedded')['files']
        for a_file in files:
            print 'file info', a_file['accession']
            f_resp = ff_utils.get_metadata(a_file['accession'], connection=ff)
            if f_resp.get('related_files'):
                print 'paired with', [i['file'].split('/')[2] for i in f_resp['related_files'] if i['relationship_type'] == 'paired with']
            else:
                print('no pair')
            print
    print
    print
    
```
