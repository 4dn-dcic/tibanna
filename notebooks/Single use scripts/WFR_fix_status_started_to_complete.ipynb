```python
from core.utils import Tibanna
from core import ff_utils

env = 'fourfront-webprod'
tibanna = Tibanna(env=env)

file_not_found = 0
deleted = 0
success = 0
patch_problem = 0
ff = ff_utils.fdn_connection(key=tibanna.ff_keys)

item_type = "WorkflowRunAwsem"
date = "2018-02-04"
wf_title = 'FastQC'
get_object = "/search/?limit=all&q=date_created%3A%3E%3D{date}&type={type}&workflow.title={wf}".format(date=date, type=item_type, wf=wf_title)
print(get_object)
wfrs_resp = ff_utils.get_metadata(get_object, connection=ff)['@graph']
print(len(wfrs_resp))
```

```python



my_wf = 'bwa-mem 0.2.5'
inp_file_args = ['fastq1', 'fastq2']
out_file_args = ['out_bam']



all_wfrs = [i for i in wfrs_resp if i['display_title'].startswith(my_wf)]  
print len(all_wfrs), 'runs'

counter = 0
for a_wfr in all_wfrs:
    inp_size = 0
    out_size = 0
    #skip deleted
    if a_wfr['status'] == 'deleted':
        continue
    # get input values
    for inp_file in a_wfr['input_files']:
        f_size = 0
        if inp_file['workflow_argument_name'] in inp_file_args:
            try:
                f_size = ff_utils.get_metadata(inp_file['value'], connection=ff)['file_size']
                inp_size += f_size          
            except:
                print 'problem'
                
    # get output values   
    for out_file in a_wfr['output_files']:
        f_size = 0
        if out_file['workflow_argument_name'] in out_file_args:
            try:
                f_size = ff_utils.get_metadata(out_file['value'], connection=ff)['file_size']
                out_size += f_size          
            except:
                print 'problem'
    print inp_size, ',' , out_size

```
